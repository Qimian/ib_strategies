```python
import requests
import json
import time

# 禁用SSL警告（因为本地网关使用自签名证书）
import urllib3
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

# 配置
BASE_URL = "http://localhost:6688/v1/api/"
AAPL_CONID = "265598"  # 苹果的ConId，你可以通过之前的查询函数获得

def get_market_snapshot(conid):
    """获取单个合约的市场快照"""
    endpoint = "iserver/marketdata/snapshot"
    
    # 参数说明：
    # conids: 合约ID，可以同时请求多个，用逗号分隔
    # fields: 想要获取的字段ID，例如 31(最新价), 55(股票代码), 84(最高价), 86(成交量)
    params = {
        "conids": conid,
        "fields": "31,55,84,86,88,89"  # 88买价, 89卖价
    }
    
    try:
        response = requests.get(
            url=f"{BASE_URL}{endpoint}",
            params=params,
            verify=False
        )
        response.raise_for_status()
        data = response.json()
        
        print(json.dumps(data, indent=2))
        return data
        
    except Exception as e:
        print(f"获取快照失败: {e}")
        return None

# 使用示例：轮询5次，每次间隔2秒
if __name__ == "__main__":
    for i in range(5):
        print(f"\n--- 请求 {i+1} ---")
        snapshot = get_market_snapshot(AAPL_CONID)
        time.sleep(2)
```
